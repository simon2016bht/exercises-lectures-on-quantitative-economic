#### Now, starting with your solution to exercise 5, plot three simulated time series, one for each of the cases α = 0, α = 0.8 and α = 0.98

```python
import numpy as np 
import matplotlib.pyplot as plt

T = 200
series = np.zeros((3,T+1))
alphas = [0, 0.8, 0.98]

for i,alpha in enumerate(alphas):
    for j in range(T):
        disturbance = np.random.normal()
        curr_value = alpha * series[i][j] + disturbance
        series[i][j+1] = curr_value

for i,alpha in enumerate(alphas):
    plt.plot(series[i], label=f'alpha = {alpha}')

plt.legend()
plt.show()
```
