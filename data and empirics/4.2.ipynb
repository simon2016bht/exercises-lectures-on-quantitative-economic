Exercise 2

Filter the above dataframe to only include employment as a percentage of ‘active population’

Create a grouped boxplot using seaborn of employment rates in 2015 by age group and sex

Hint: GEO includes both areas and countries

```python
data.head()
data.sort_index(axis=1)
```

```python
geo_list = data.columns.get_level_values('GEO').unique()
```

```python
geo_list
```

```python

ll = [x for x in geo_list if not x.startswith('Euro')]
data = data[ll]
```

```python
data = data.xs(('Percentage of total population', 'Active population'), axis=1, level=('UNIT', 'INDIC_EM'))
```

```python
data.head()
```

```python
data.drop('Total', axis=1, level='SEX', inplace=True)
```

```python
data.index = pd.to_datetime(data.index)
```

```python
box = data['2015'].unstack().reset_index()
sns.boxplot(x="AGE", y=0, hue='SEX', data=box, palette=('husl'))
```
