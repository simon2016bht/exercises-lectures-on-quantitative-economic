```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta = np.linspace(0,10,100)
y = np.sin(theta)
plt.plot(theta,y)
plt.show()
```

```python
thetas = np.linspace(0,2,10)
x = np.linspace(0,5,100)
for theta in thetas:
    y = np.cos(theta*np.pi*x)*np.exp(-x)
    plt.plot(x,y)    
```

```python
fix, ax = plt.subplots()
thetas = np.linspace(0,2,100)
x = np.linspace(0,5,100)
for theta in thetas:
    y = np.cos(theta*np.pi*x)*np.exp(-x)
    ax.plot(x,y,label=theta)  

#ax.legend()    
plt.show()
```
